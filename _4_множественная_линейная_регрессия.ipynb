4.	Разделите все наблюдения на обучающую и тестовую выборки в соотношении 80-20.

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2)
```

5.	Для целевой и каждого из признаков постройте парную регрессионную модель. Визуализируйте линию регрессии и исходные данные на графике. При помощи коэффициентов оцените силу связи.

```python
for reg_name in X_train.columns:
    reg = linear_model.LinearRegression()

    x_train = np.array(X_train[reg_name]).reshape(-1, 1)
    x_test = np.array(X_test[reg_name]).reshape(-1, 1) 
                                                       
    reg.fit(x_train, y_train)
    y_predict = reg.predict(x_test)
    y_predict_train = reg.predict(x_train)

    print('Feature  ', reg_name) 
    print('Training data R^2:', reg.score(x_train, y_train))

    print('Test data MSE:', mean_squared_error(y_predict, y_test))
    print('Training data MSE:', mean_squared_error(y_predict_train, y_train))

    print('Test data MAE:', mean_absolute_error(y_predict, y_test))
    print('Training data MAE:', mean_absolute_error(y_predict_train, y_train))

    print('Estimated intercept:', reg.intercept_[0])
    print('Estimated slope coef:', reg.coef_[0]) 

    plt.scatter(x_test, y_test) 
    plt.plot(x_test, y_predict, 'r') 
    plt.show()
```

6.	Постройте модель множественной линейной регрессии целевой переменной от всех показателей. Оцените качество регрессии, сделайте вывод.

```python
reg = linear_model.LinearRegression()
reg.fit(X_train, y_train)
y_predict = reg.predict(X_test)
y_predict_train = reg.predict(X_train)

print('Training data R^2:', reg.score(X_train, y_train))

print('Test data MSE:', mean_squared_error(y_predict, y_test))
print('Training data MSE:', mean_squared_error(y_predict_train, y_train))

print('Test data MAE:', mean_absolute_error(y_predict, y_test))
print('Training data MAE:', mean_absolute_error(y_predict_train, y_train))

print('Estimated intercept:', reg.intercept_[0])
print('Estimated slope coef:', reg.coef_[0])
```

## Дополнительные задания

1.	Для целевой переменной и каждого из признаков постройте полиномиальную регрессию второго, третьего и десятого порядков. Сделайте вывод о значимости регрессии.

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
def each_pair_poli_regression(X_train, X_test, y_train, y_test, n):
    for reg_name in X_train.columns:
        poly_reg = PolynomialFeatures(degree = n)
        X_poly_train = poly_reg.fit_transform(np.array(X_train[reg_name]).reshape(-1,1))
        X_poly_test = poly_reg.transform(np.array(X_test[reg_name]).reshape(-1,1))
        
        pol_reg = linear_model.LinearRegression()
        pol_reg.fit(X_poly_train, y_train)
        
        y_predict = pol_reg.predict(X_poly_test)
        y_predict_train = pol_reg.predict(X_poly_train)
        
        print('Feature  ', reg_name)
        
        print('Training data R^2:', pol_reg.score(X_poly_train, y_train))
        
        print('Test data MSE:', mean_squared_error(y_predict, y_test))
        print('Training data MSE:', mean_squared_error(y_predict_train, y_train))
        
        print('Test data MAE:', mean_absolute_error(y_predict, y_test))
        print('Training data MAE:', mean_absolute_error(y_predict_train, y_train))
        
        print('Estimated intercept:', pol_reg.intercept_[0])
        print('Estimated slope coef:', pol_reg.coef_[0])
        
        plot_x_test = np.linspace(X_test[reg_name].min(), X_test[reg_name].max(), 100)
        plot_poly_x_test = poly_reg.fit_transform(plot_x_test.reshape(-1,1))
        plt.scatter(X_test[reg_name], np.array(y_test).reshape(1,len(y_test)), color='black')
        plt.plot(plot_x_test, pol_reg.predict(plot_poly_x_test), color='red', label = f'Признак {reg_name} with deg = {n}')
        plt.show()
```

```python
each_pair_poli_regression(X_train, X_test, y_train, y_test, 2)
```

```python
each_pair_poli_regression(X_train, X_test, y_train, y_test, 3)
```
